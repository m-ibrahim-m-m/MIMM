```python
Model = LinearRegression()
isinstance(Model,LinearRegression) # Just for check
```

```python
x = df['YearsExperience']
x = np.array(x).reshape(-1,1)
x
```

```python
y = df['Salary']
y
```

```python
Model.fit(x ,y )
```

```python
b = Model.intercept_
b
```

```python
a = Model.coef_
a
```

## **y_ = Model.coef_ * x +Model.intercept_**

```python
y_perdict = Model.predict(x)
y_perdict
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_percentage_error,mean_absolute_error
```

```python
mean_squared_error(y_perdict,y)
```

```python
mean_absolute_error(y_perdict,y)
```

```python
mean_absolute_percentage_error(y_perdict,y)*100
```
